```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_train = df_train.drop('id', axis=1)
df_test = df_test.drop('id', axis=1)
```

```python
df_train.info()
```

```python
df_test.info()
```

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
df_train['Brand'].value_counts()
```

```python
# df_train['Brand'] = df_train['Brand'].apply(lambda x: np.random.choice(df_train['Brand'].value_counts().index) if pd.isna(x) else x)
# df_test['Brand'] = df_test['Brand'].apply(lambda x: np.random.choice(df_test['Brand'].value_counts().index) if pd.isna(x) else x)
```

```python
df_train['Brand'] = df_train['Brand'].fillna('Adidas')
df_test['Brand'] = df_test['Brand'].fillna('Adidas')
```

```python
df_train['Material'].value_counts()
```

```python
df_train['Material'] = df_train['Material'].fillna('Polyester')
df_test['Material'] = df_test['Material'].fillna('Polyester')
```

```python
df_train['Size'].value_counts()
```

```python
df_train['Size'] = df_train['Size'].fillna('Medium')
df_test['Size'] = df_test['Size'].fillna('Medium')
```

```python
df_train['Laptop Compartment'].value_counts()
```

```python
df_train['Laptop Compartment'] = df_train['Laptop Compartment'].fillna('Yes')
df_test['Laptop Compartment'] = df_test['Laptop Compartment'].fillna('Yes')
```

```python
df_train['Waterproof'] = df_train['Waterproof'].fillna('Yes')
df_test['Waterproof'] = df_test['Waterproof'].fillna('Yes')
```

```python
df_train['Style'].value_counts()
```

```python
df_train['Style'] = df_train['Style'].fillna('Messenger')
df_test['Style'] = df_test['Style'].fillna('Messenger')
```

```python
df_train['Color'].value_counts()
```

```python
df_train['Color'] = df_train['Color'].fillna('Pink')
df_test['Color'] = df_test['Color'].fillna('Pink')
```

```python
df_train['Weight Capacity (kg)'] = df_train['Weight Capacity (kg)'].fillna(df_train['Weight Capacity (kg)'].median())
df_test['Weight Capacity (kg)'] = df_test['Weight Capacity (kg)'].fillna(df_test['Weight Capacity (kg)'].median())
```

```python
df_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
for i in df_test.columns:
    if df_train[i].dtype == object:
        if len(set(df_test[i]).difference(set(df_train[i]))) == 0:
            df_train[i] = label.fit_transform(df_train[i])
            df_test[i] = label.transform(df_test[i])
        else:
            print(i)
```

```python
df_train.info()
```

```python
df_test.info()
```

```python
X_train = df_train.drop('Price', axis=1)
y_train = df_train['Price']
X_test = df_test.copy()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
model = DecisionTreeRegressor()
scaler = StandardScaler()
```

```python
pipe = Pipeline([('scaler', scaler), ('model', model)])
```

```python
params = {
    'model__criterion': ['squared_error', 'friedman_mse'],
    'model__max_depth': [5, 8],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [1, 2, 5],
    'model__max_features': [None, 'sqrt', 'log2']
}
```

```python
final_model = GridSearchCV(pipe, params, cv=3, verbose=2)
```

```python
final_model.fit(X_train, y_train)
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub['Price'] = y_pred
```

```python
df_sub.to_csv('sub.csv', index=False, index_label=False)
```
